### The Metric: Pearson Correlation Coefficient(r): This measures the linear relationship between two variables.
### +1: Perfect Positive Correlation (As X goes up, Y goes up).
### -1: Perfect Negative Correlation (As X goes up, Y goes down).
### 0: No linear relationship (Random noise).
#### Warning ⚠️: "Correlation does not imply Causation." Just because ice cream sales and shark attacks both go up in summer (correlation), doesn't mean ice cream causes shark attacks (causation). They are likely both driven by a third variable: Temperature.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Generate Synthetic Data
x = np.linspace(0, 10, 100)

# Positive Correlation (y ~ x)
y_pos = x + np.random.normal(0, 1, 100)

# Negative Correlation (y ~ -x)
y_neg = -x + np.random.normal(0, 1, 100)

# No Correlation (y is random)
y_none = np.random.normal(0, 1, 100)

# 2. Plotting side-by-side
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Positive
sns.scatterplot(x=x, y=y_pos, ax=axes[0], color='green')
axes[0].set_title(f'Positive Correlation\nr ≈ {np.corrcoef(x, y_pos)[0,1]:.2f}')

# Plot 2: Negative
sns.scatterplot(x=x, y=y_neg, ax=axes[1], color='red')
axes[1].set_title(f'Negative Correlation\nr ≈ {np.corrcoef(x, y_neg)[0,1]:.2f}')

# Plot 3: None
sns.scatterplot(x=x, y=y_none, ax=axes[2], color='blue')
axes[2].set_title(f'No Correlation\nr ≈ {np.corrcoef(x, y_none)[0,1]:.2f}')

plt.show()
```
