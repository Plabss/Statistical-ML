```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
import numpy as np

# 1. Load Data
digits = load_digits()
X = digits.data
y = digits.target

# 2. Define Model
model = RandomForestClassifier(n_estimators=40, random_state=42)

# 3. Run K-Fold Cross-Validation (K=10)
# This trains the model 10 times on different splits
scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')

# 4. Results
print(f"Scores for each run: {scores}")
print(f"Average Accuracy: {np.mean(scores) * 100:.2f}%")
print(f"Standard Deviation: {np.std(scores) * 100:.2f}%")
```
