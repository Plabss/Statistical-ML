```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 1. Load Data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# 2. Define the Model
model = SVC()

# 3. Define the Grid (The settings we want to test)
# C: Strictness of the boundary (High C = Strict, Low C = Loose)
# kernel: The math used to draw the boundary
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly']
}

# 4. Run Grid Search
# cv=5 means "Cross Validation" (we'll explain this next!)
grid = GridSearchCV(model, param_grid, refit=True, verbose=2, cv=5)
grid.fit(X_train, y_train)

# 5. Results
print(f"Best Parameters Found: {grid.best_params_}")
print(f"Best Accuracy: {grid.best_score_ * 100:.2f}%")

# 6. Evaluate on Test Data
from sklearn.metrics import classification_report
grid_predictions = grid.predict(X_test)
print(classification_report(y_test, grid_predictions))
```
