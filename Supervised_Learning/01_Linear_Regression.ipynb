```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Generate Synthetic Data (Experience vs Salary)
np.random.seed(42)
X = 2 * np.random.rand(100, 1) # Years of Experience (0 to 2 years)
y = 4 + 3 * X + np.random.randn(100, 1) # Salary (y = 4 + 3x + noise)

# 2. Train the Model
model = LinearRegression()
model.fit(X, y) # This is where the machine "learns" the line

# 3. Make Predictions
X_new = np.array([[0], [2]])
y_predict = model.predict(X_new)

# 4. Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X_new, y_predict, color='red', linewidth=3, label='Prediction (Best Fit Line)')
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in $10k)")
plt.title(f"Linear Regression Fit\nBias (0): {model.intercept_[0]:.2f}, Weight (1): {model.coef_[0][0]:.2f}")
plt.legend()
plt.show()

# 5. Check Error
print(f"Model Intercept (Bias): {model.intercept_}")
print(f"Model Coefficient (Weight): {model.coef_}")
```
