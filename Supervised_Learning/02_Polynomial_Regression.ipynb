```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import PolynomialFeatures

# 1. Generate Non-Linear Data (Quadratic)
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1) # y = 0.5x^2 + x + 2 + noise

# 2. Transform Features (Add x^2)
# We convert our single feature X into X and X^2
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# 3. Train Linear Regression on the Transformed Data
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# 4. Visualization
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.6, label='Actual Data')
plt.plot(X_new, y_new, color='red', linewidth=3, label='Polynomial Prediction')
plt.title("Polynomial Regression (Degree 2)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# 5. Check Coefficients
print(f"Intercept: {lin_reg.intercept_}")
print(f"Coefficients: {lin_reg.coef_}")
```

```python
from sklearn.preprocessing import PolynomialFeatures

# 1. Generate Non-Linear Data (Quadratic)
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1) # y = 0.5x^2 + x + 2 + noise

# 2. Transform Features (Add x^2)
# We convert our single feature X into X and X^2
poly_features = PolynomialFeatures(degree=30, include_bias=False)
X_poly = poly_features.fit_transform(X)

# 3. Train Linear Regression on the Transformed Data
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# 4. Visualization
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.6, label='Actual Data')
plt.plot(X_new, y_new, color='red', linewidth=3, label='Polynomial Prediction')
plt.title("Polynomial Regression (Degree 2)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# 5. Check Coefficients
print(f"Intercept: {lin_reg.intercept_}")
print(f"Coefficients: {lin_reg.coef_}")
```

### **Overfitting: The model is memorizing the noise, not learning the pattern.**

### Model Complexity and the BiasVariance Tradeoff

**Degree 1 (Underfitting):**  
- The model is too simple to capture the underlying trend  
- Results in **high bias**

**Degree 10 (Overfitting):**  
- The model captures noise and randomness instead of the true signal  
- Fails to generalize to new, unseen data  
- Results in **high variance**

---

### Key Takeaway

Finding the right balance between underfitting and overfitting is known as the  
**BiasVariance Tradeoff**  a **fundamental concept** frequently discussed in data science and machine learning interviews.
